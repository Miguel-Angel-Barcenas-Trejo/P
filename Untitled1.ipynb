```python
#Implementar un árbol desde cero
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

root = Node(10)

root.left = Node(34)
root.right = Node(89)
root.left.left = Node(45)
root.left.right = Node(50)



#Recorrido posterior al pedido
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data)

# Llamada a la función postorder
print("Recorrido en postorder:")
postorder(root)
```

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

root = Node(10)

root.left = Node(34)
root.right = Node(89)
root.left.left = Node(45)
root.left.right = Node(50)

# Recorrido de pedidos anticipados
def preorder(node):
    if node:
        print(node.data)
        preorder(node.left)
        preorder(node.right)

# Llamada a la función preorder
print("Recorrido en preorden:")
preorder(root)
```

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

root = Node(10)

root.left = Node(34)
root.right = Node(89)
root.left.left = Node(45)
root.left.right = Node(50)

#Recorrido en orden
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data)
        inorder(node.right)

# Llamada a la función orden
print("Recorrido en preorden:")
inorder(root)
```

```python
# Clase llamada Nodo
class Nodo:
    # Constructor con los argumentos
    def __init__(self, value=None, izq=None, der=None):
        # Punteros
        self.value = value
        self.izq = izq
        self.der = der

    # Retornar el valor
    def __str__(self):
        return str(self.value)

# Clase del árbol binario
class ArbolBinario:
    # Método constructor y el atributo raiz
    def __init__(self):
        self.raiz = None

    # Creamos la función agregar
    def agregar(self, elemento):
        if self.raiz == None:
            self.raiz = elemento
        else:
            aux = self.raiz
            padre = None
            while aux != None:
                padre = aux
                if int(elemento.value) >= int(aux.value):
                    aux = aux.der
                else:
                    aux = aux.izq
            if int(elemento.value) >= int(padre.value):
                padre.der = elemento
            else:
                padre.izq = elemento

    # Se crea método mostrar el preorden
    def preorden(self, elemento):
        if elemento != None:
            print(elemento)
            self.preorden(elemento.izq)
            self.preorden(elemento.der)

    # Se crea método mostrar el postorden (recursividad)
    def postorden(self, elemento):
        if elemento != None:
            self.postorden(elemento.izq)
            self.postorden(elemento.der)
            print(elemento)

    # Se crea método mostrar el inorden
    def inorden(self, elemento):
        if elemento != None:
            self.inorden(elemento.izq)
            print(elemento)
            self.inorden(elemento.der)

    # Se crea función que permite obtener la raíz
    def getRaiz(self):
        return self.raiz

# Creamos un menú
if __name__ == "__main__":
    # Llamamos el objeto de árbol binario
    ab = ArbolBinario()

    while True:
        # Opciones del menú
        print("Arboles_Binarios\n"+
             "1. Agregar\n"+
             "2. Preorden\n"+
             "3. Postorden\n"+
             "4. Inorden\n"+
             "5. Salir")

        num = input("Ingrese la opcion: ")

        if num == "1":
            value = input("Ingrese el valor: ")
            nod = Nodo(value)
            ab.agregar(nod)
        elif num == "2":
            print("Preorden")
            ab.preorden(ab.getRaiz())
        elif num == "3":
            print("Postorden")
            ab.postorden(ab.getRaiz())
        elif num == "4":
            print("Inorden")
            ab.inorden(ab.getRaiz())
        elif num == "5":
            exit()
```
